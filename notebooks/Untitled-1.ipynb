```python
import VASPsol as vs
import numpy as np
import pandas as pd
import os
import COSMO_TL as ctl
from dask.distributed import Client, LocalCluster, progress
import dask
```

```python
ctl.gl.get_tr_sigma_profile('test3014_0_1')
```

```python
cluster = LocalCluster(n_workers=4)
client = Client(cluster)
client
```

```python
# get all dirs in the current directory that have a directory called "VAC" in them
run_dirs = [os.path.join(os.getcwd(), d) for d in os.listdir(os.getcwd()) if os.path.isdir(os.path.join(os.getcwd(), d)) and os.path.isdir(os.path.join(os.getcwd(), d, "VAC"))]
run_dirs
```

```python
jobs = [client.submit(vs.data, i) for i in run_dirs]
progress(jobs)
```

```python
data = client.gather(jobs)
df = pd.concat([i.ml_df for i in data])
df
```

```python
#default vaspsol params
NC_K_default = 0.0025
SIGMA_K_default = 0.6
TAU_default = 0.000525
```

```python
# default df
default_df = df[(df['NC_K'] == NC_K_default) & (df['SIGMA_K'] == SIGMA_K_default) & (df['TAU'] == TAU_default)]
default_df
```

```python
ctl.GammaLoaders
```

```python
default_df['SoluteName'].iloc[0].upper()
# create a try function to get the sigma profile
def get_sig(solute_name):
    try:
        return ctl.GammaLoaders.get_sigma_profile(solute_name)
    except:
        return None
```

```python
sigs = [client.submit(get_sig, i) for i in default_df['SoluteName'].str.upper()]
sigs
```

```python
sigs = []
for idx, i in enumerate(default_df['SoluteName'].str.upper()):
    try:
        sigs.append(ctl.GammaLoaders.get_sigma_profile(i).compute())
    except:
        try:
            sigs.append(ctl.GammaLoaders.get_sigma_profile(default_df['Filehandle'].iloc[i]).compute())
        except:
            pass
final_sigs = [i for i in sigs if i is not None]
len(final_sigs)
```
