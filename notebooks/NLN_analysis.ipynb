```python
import VASPsol as vs
import os
import numpy as np
import pandas as pd
```

```python
files = [os.path.abspath('../calculations/'+f) for f in os.listdir('../calculations/') if f]
# get only the directories
files = [f for f in files if os.path.isdir(f)]
files
```

```python
from dask.distributed import Client, progress
client = Client(n_workers=4, threads_per_worker=1, memory_limit='2GB')
client
```

```python
jobs = [client.submit(vs.data, f) for f in files]
progress(jobs)
```

```python
data = client.gather(jobs)
df = pd.concat([i.ml_df for i in data]).reset_index(drop=True)
df
```

```python
nln = df[df['calc_type'] == 'non-li']
nln
```

```python
nln['error'] = nln['error'].abs()
nln = nln[nln['error'] < 10]
nln['error'].describe()
```
