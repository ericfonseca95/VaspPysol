```python
data = client.gather(data_jobs)
df = pd.concat([data.ml_df for data in data])
df
```

```python
df2 = df[df['SoluteName'].isin(names)]
df2
```

```python
# get the FileHandles for the data
filenames = np.unique(df2['FileHandle'].values)
# lets make a grid of values of NC_K, SIGMA_K to loop over. we are going to stick with the default value of TAU
TAU = 5.25e-4
NC_K = np.linspace(1e-3, 4e-3, 9)
SIGMA_K = np.linspace(0.2, 1, 9)
NC_K, SIGMA_K 
```

```python
jobs = [client.submit(vs.calculate_solvent, file, dict_of_additional_tags={'NC_K':n, 
                                                                           'SIGMA_K':s, 
                                                                           'TAU':TAU,
                                                                           'NSW':0,
                                                                           'NELM':60}) for file in filenames for n in NC_K for s in SIGMA_K]
progress(jobs)
```
